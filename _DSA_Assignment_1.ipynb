# 5. Perform meaningful visualizations using the dataset. 
# Bring at least 3 visualizations.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Box plot

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=df)
plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SW',data=df)
plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='PL',data=df)
plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PW',data=df)
plt.show()
```

 # Scatter plot

```python
plt.figure(figsize=(8,9))
#plt.scatterplot(df['PL'],df['PW'],s=15)
plt.title("PL-PW ")
plt.xlabel('PL')
plt.ylabel('PW')
sns.scatterplot(x='PL', y='PW',hue='Classification', data=df,s=30 )
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
plt.figure(figsize=(8,9))

plt.title("SL-SW ")
plt.xlabel('SL')
plt.ylabel('SW')
sns.scatterplot(x='SL', y='SW',hue='Classification', data=df,s=30)
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
sns.countplot(x='Classification', data=df, )
plt.show()
```
